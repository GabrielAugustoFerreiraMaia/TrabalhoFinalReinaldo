```python
import pandas as pd
import matplotlib.pyplot as plt

clientes_df = pd.read_csv("/content/olist_customers(Clientes)_dataset.csv")
loclizacao_df = pd.read_csv("/content/olist_geolocation(GeoLocalizacao)_dataset.csv")
itensPedido_df = pd.read_csv("/content/olist_order_items(itens_pedido)_dataset.csv")
pagamento_df = pd.read_csv("/content/olist_order_payments(pagamento_pedido)_dataset.csv")
avaliacaoPedido_df = pd.read_csv("/content/olist_order_reviews(avaliacao_pedido)_dataset.csv")
pedidos_df = pd.read_csv("/content/olist_orders(Pedidos)_dataset.csv")
produtos_df = pd.read_csv("/content/olist_products(Produtos)_dataset.csv")
vendedores_df = pd.read_csv("/content/olist_sellers(Vendedores)_dataset.csv")

clientes_df
```

```python
quant_clientes = clientes_df['customer_state'].value_counts()

quant_clientes
```

```python

quant_clientes.plot(kind='bar', figsize=(10, 6))
plt.title("Origem dos Clientes")
plt.xlabel("Estado")
plt.ylabel("Quantidade de Clientes")
plt.show()
```

```python
#Fazendo Merge nos df itens pedido e 
join_df=pd.merge(itensPedido_df,produtos_df, on='product_id')
join_df
```

```python
categorias_populares = join_df['product_category_name'].value_counts()
categorias_populares

top_categorias = categorias_populares.head(20)
```

```python
top_categorias.plot(kind='bar', figsize=(10, 6))
plt.title("Categorias Populares")
plt.xlabel("Categorias")
plt.ylabel("Produtos Vendidos")
plt.show()
```
